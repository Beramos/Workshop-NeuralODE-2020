A simple example: **One input** and **no activation** function. You actually know this as **linear regression**.

```julia
using Plots
```

```julia
a = -0.3;
b = 7.5;
```

```julia
f(x) = a*x+b
```

```julia
f(2.0)
```

```julia
names = ["mouse", "rat", "rabbit", "dog", "human"]
mass = [25.0, 200.0, 2000.0, 30000.0, 70000.0]
pulseRate = [670.0, 420.0, 205.0, 85.0, 72.0];
```

```julia
Y₁ = log.(pulse);
X₁ = log.(mass);
```

```julia
scatter(X₁, Y₁,series_annotations = text.(names,:bottom),
    label="", xlabel="log mass (g)", ylabel="log pulse rate (BPM)", 
    ylims=(4,7), xlims=(2,12)) 

plot!(X₁, f.(X₁), color=:grey, lw=2, ls=:dash, label="f(x) = $(a)*x+$(b)")
```

<div class="alert alert-block alert-info">
    <b>Neural networks are a non-linear combination of linear models.</b>
    The values of the linear models pass through a non-linear activation function generating the input values for the next layer.
</div>

```julia
a₁, a₂ = -2, 10
b₁, b₂, b₃ = -1, -3, 3;
```

```julia
f₁(x) = tanh(f(x))
f₂(x) = (f₁(x) + a₁*f₁(x+b₁) + f₁(x+b₂))
f₃(x) = tanh(a₂*f₂(x)+a₁*f₂(x+b₃));
```

```julia
X = 20:0.1:40
plot(X, f₁.(X), 
    label="f₁(x) = tanh($(a)*x+$(b))",xlabel = "X", ylabel="f(X)",
    lw = 2, legend=:bottomright)
plot!(X, f₂.(X), label = "f₂(x) =f₁(x) + a₁*f₁(x+b₁) + f₁(x+b₂)")
plot!(X, f₃.(X), label = "f₃(x) = tanh(a₂*f₂(x)+a₁*f₂(x+b₃))")
```

The final equation for f<sub>3</sub>(x) is,

$$f_3(x)= \mathrm{tanh}\left(a_2\,(\mathrm{tanh}(-0.3\,x + 7.5) + a₁\,\mathrm{tanh}(-0.3\,x + 7.5) + \mathrm{tanh}(-0.3\,x + 7.5))) + a_1\,(\mathrm{tanh}(-0.3\,x + 7.5) + a₁\,\mathrm{tanh}(-0.3\,x + 7.5) + \mathrm{tanh}(-0.3\,x + 7.5))\right)$$

and is actually a simple, 3-layer neural-network where the non-linear function (activation function) is a hyperbolic tangens.
